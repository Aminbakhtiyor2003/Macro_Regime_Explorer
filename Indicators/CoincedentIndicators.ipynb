```python
import requests
import pandas as pd

# Real GDP Growth (Percent Change, Quarterly)
series_id_gdp = 'GDP'
units_gdp = 'pch'              # Percent change (QoQ)
frequency_gdp = 'q'            # Quarterly frequency
start_date_gdp = '1947-01-01'
end_date_gdp = '2025-03-27'

# URL and parameters
url_gdp = "https://api.stlouisfed.org/fred/series/observations"
params_gdp = {
    'series_id': series_id_gdp,
    'api_key': api_key,
    'file_type': 'json',
    'units': units_gdp,
    'frequency': frequency_gdp,
    'observation_start': start_date_gdp,
    'observation_end': end_date_gdp
}

# Make the request
response_gdp = requests.get(url_gdp, params=params_gdp)

# Format response
if response_gdp.status_code == 200:
    data_gdp = response_gdp.json()
    df_gdp = pd.DataFrame(data_gdp['observations'])

    df_gdp['date'] = pd.to_datetime(df_gdp['date'])
    df_gdp['value'] = pd.to_numeric(df_gdp['value'], errors='coerce')
    df_gdp = df_gdp[['date', 'value']].set_index('date')
    df_gdp = df_gdp.dropna()

    df_gdp.rename(columns={'value': 'GDP Growth (% QoQ)'}, inplace=True)
    df_gdp.index.name = "Date"

    print(df_gdp.head(10))

else:
    print('Failed to retrieve GDP data.')
    print('Status Code:', response_gdp.status_code)
    print(response_gdp.text)

```

```python
# Nonfarm Payrolls (Total Employment) - PAYEMS
series_id_payems = 'PAYEMS'
units_payems = 'lin'             # Raw employment level (in thousands)
frequency_payems = 'm'
start_date_payems = '1948-01-01'
end_date_payems = '2025-03-27'

# URL and parameters
url_payems = "https://api.stlouisfed.org/fred/series/observations"
params_payems = {
    'series_id': series_id_payems,
    'api_key': api_key,
    'file_type': 'json',
    'units': units_payems,
    'frequency': frequency_payems,
    'observation_start': start_date_payems,
    'observation_end': end_date_payems
}

# Make the request
response_payems = requests.get(url_payems, params=params_payems)

# Format response
if response_payems.status_code == 200:
    data_payems = response_payems.json()
    df_payems = pd.DataFrame(data_payems['observations'])

    df_payems['date'] = pd.to_datetime(df_payems['date'])
    df_payems['value'] = pd.to_numeric(df_payems['value'], errors='coerce')
    df_payems = df_payems[['date', 'value']].set_index('date')
    df_payems = df_payems.dropna()

    # Calculate Year-over-Year % Change
    df_payems['NonfarmPayrolls'] = df_payems['value'].pct_change(periods=12) * 100
    df_payems = df_payems[['NonfarmPayrolls']].dropna()
    df_payems.index.name = "Date"

    print(df_payems.head(10))

else:
    print('Failed to retrieve Nonfarm Payrolls data.')
    print('Status Code:', response_payems.status_code)
    print(response_payems.text)
```

```python
# Advance Retail Sales: Retail Trade and Food Services
series_id_retail = 'RSAFS'
units_retail = 'lin'         
frequency_retail = 'm'        # Monthly
start_date_retail = '1992-01-01'
end_date_retail = '2025-03-27'

# URL and parameters
url_retail = "https://api.stlouisfed.org/fred/series/observations"
params_retail = {
    'series_id': series_id_retail,
    'api_key': api_key,
    'file_type': 'json',
    'units': units_retail,
    'frequency': frequency_retail,
    'observation_start': start_date_retail,
    'observation_end': end_date_retail
}

#  Make the API request
response_retail = requests.get(url_retail, params=params_retail)

#  Format and transform the data
if response_retail.status_code == 200:
    data_retail = response_retail.json()
    df_retail = pd.DataFrame(data_retail['observations'])

    # Parse dates and numeric values
    df_retail['date'] = pd.to_datetime(df_retail['date'])
    df_retail['value'] = pd.to_numeric(df_retail['value'], errors='coerce')
    df_retail = df_retail[['date', 'value']].set_index('date').dropna()

    # Compute Year-over-Year Percent Change
    df_retail['RetailSales'] = df_retail['value'].pct_change(periods=12) * 100

    # Keep only YoY column
    df_retail = df_retail[['RetailSales']].dropna()

    # Preview the data
    print(df_retail.head())

else:
    print('Failed to retrieve Retail Sales data.')
    print('Status Code:', response_retail.status_code)
    print(response_retail.text)
```

```python
# Merge all Coincident Indicators
df_coincident_indicators = pd.concat([
    df_gdp,
    df_payems,
    df_retail,
    df_indpro,
    df_pce,
    df_income,
    df_corp
], axis=1)

# Find the first date where all coincident indicators have data
first_valid_date_coincident = df_coincident_indicators.dropna().index.min()

# Trim data to start from that date
df_coincident_indicators = df_coincident_indicators.loc[first_valid_date_coincident:]

# Forward fill to handle gaps across monthly/quarterly data
df_coincident_indicators = df_coincident_indicators.ffill()

# Format all numeric columns to "xx.xx"
for col in df_coincident_indicators.columns:
    df_coincident_indicators[col] = df_coincident_indicators[col].map(lambda x: f"{x:.2f}")

# Reset index and add Date column
df_coincident_indicators = df_coincident_indicators.reset_index()
df_coincident_indicators.rename(columns={'index': 'Date'}, inplace=True)

# Preview the table
print(df_coincident_indicators.head())
```
