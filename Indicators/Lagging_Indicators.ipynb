```python
import requests
import pandas as pd


# Core CPI (% MoM) - Excludes Food & Energy
series_id_corecpi = 'CPILFESL'
units_corecpi = 'pch'           # Percent change (Month-over-Month)
frequency_corecpi = 'm'         # Monthly
start_date_corecpi = '1957-01-01'
end_date_corecpi = '2025-03-27'

# URL and parameters
url_corecpi = "https://api.stlouisfed.org/fred/series/observations"
params_corecpi = {
    'series_id': series_id_corecpi,
    'api_key': api_key,
    'file_type': 'json',
    'units': units_corecpi,
    'frequency': frequency_corecpi,
    'observation_start': start_date_corecpi,
    'observation_end': end_date_corecpi
}

# Make the request
response_corecpi = requests.get(url_corecpi, params=params_corecpi)

# Format response
if response_corecpi.status_code == 200:
    data_corecpi = response_corecpi.json()
    df_corecpi = pd.DataFrame(data_corecpi['observations'])

    df_corecpi['date'] = pd.to_datetime(df_corecpi['date'])
    df_corecpi['value'] = pd.to_numeric(df_corecpi['value'], errors='coerce')
    df_corecpi = df_corecpi[['date', 'value']].set_index('date')
    df_corecpi = df_corecpi.dropna()

    # Rename column
    df_corecpi.rename(columns={'value': 'Core CPI (% MoM)'}, inplace=True)
    df_corecpi.index.name = "Date"

    print(df_corecpi.head(10))

else:
    print('Failed to retrieve Core CPI data.')
    print('Status Code:', response_corecpi.status_code)
    print(response_corecpi.text)
```

```python
import requests
import pandas as pd

# Credit Spread: Baa Corporate Bond Yield minus 10Y Treasury Yield
series_id_credit = 'BAA10Y'
units_credit = 'lin'           # Already a spread in percentage points
frequency_credit = 'm'         # Monthly
start_date_credit = '1986-01-01'
end_date_credit = '2025-03-27'

# URL and parameters
url_credit = "https://api.stlouisfed.org/fred/series/observations"
params_credit = {
    'series_id': series_id_credit,
    'api_key': api_key,
    'file_type': 'json',
    'units': units_credit,
    'frequency': frequency_credit,
    'observation_start': start_date_credit,
    'observation_end': end_date_credit
}

# Make the request
response_credit = requests.get(url_credit, params=params_credit)

# Format response
if response_credit.status_code == 200:
    data_credit = response_credit.json()
    df_credit_spread = pd.DataFrame(data_credit['observations'])

    df_credit_spread['date'] = pd.to_datetime(df_credit_spread['date'])
    df_credit_spread['value'] = pd.to_numeric(df_credit_spread['value'], errors='coerce')
    df_credit_spread = df_credit_spread[['date', 'value']].set_index('date')
    df_credit_spread = df_credit_spread.dropna()

    # Rename column
    df_credit_spread.rename(columns={'value': 'Credit Spread (Baa - 10Y Treasury)'}, inplace=True)
    df_credit_spread.index.name = "Date"

    print(df_credit_spread.head(10))

else:
    print('Failed to retrieve Credit Spread data.')
    print('Status Code:', response_credit.status_code)
    print(response_credit.text)
```

```python
# 1. Merge lagging indicators
df_lagging_indicators = pd.concat([
    df_corecpi,
    df_corepce,
    df_ulc,
    df_fed_debt_gdp,
    df_credit_spread
], axis=1)

# 2. Trim from first complete data row
first_valid_date_lagging = df_lagging_indicators.dropna().index.min()
df_lagging_indicators = df_lagging_indicators.loc[first_valid_date_lagging:]

# ✅ 3. Reindex to monthly before filling
monthly_index = pd.date_range(start=df_lagging_indicators.index.min(), end=df_lagging_indicators.index.max(), freq="MS")
df_lagging_indicators = df_lagging_indicators.reindex(monthly_index)

# 4. Forward fill quarterly values across the new monthly index
df_lagging_indicators = df_lagging_indicators.ffill()

# 5. Format all values to "xx.xx"
for col in df_lagging_indicators.columns:
    df_lagging_indicators[col] = df_lagging_indicators[col].map(lambda x: f"{x:.2f}")

# ✅ Optional: Set back the index name to 'Date'
df_lagging_indicators.index.name = "Date"

# 6. Preview
print(df_lagging_indicators.head())
```

```python

# 7. Save to CSV for reuse
df_lagging_indicators.to_csv("lagging_indicators.csv")
```
